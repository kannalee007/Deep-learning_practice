```python
import pandas as pd
import numpy as np
from tensorflow import keras
import pandas as pd
from sklearn.model_selection import RandomizedSearchCV
```

```python
#fine tuning the ANN
#1. build the ANN
#2. convert that ANN to machine learning model - 

def build_ann(n_hidden=1, n_neurons=30, learning_rate=0.003, input_shape=[8]):
    #build the network
    ann = keras.models.sequential()
    ann.add(keras.layers.InputLayer(input_shape=input_shape))
    for layer in range(n_hidden):
        ann.add(keras.layers.Dense(n_neurons, activation='relu'))
    ann.add(keras.layers.Dense(1))

    #compile the network
    ann.compile(loss='mse', optimizer=keras.optimizers.SGD(learning_rate=learning_rate))
    
    return ann
    
keras_reg_model = keras.wrappers.scikit_learn.KerasRegressor(build_ann)  #convert the ann to scikit learn model
params = {'n_hidden': np.arange(1, 5).tolist(), 'n_neurons': np.arange(1, 100).tolist(), 'learning_rate': [0.1, 0.01, 0.001]}
random_search_cv = RandomizedSearchCV(keras_reg_model, params, cv=3, n_iter=10)
random_search_cv.fit(X_train, y_train, epochs=100, validation_data=(X_valid, y_valid), callbacks=[keras.callbacks.EarlyStopping(patience=10)])
```
